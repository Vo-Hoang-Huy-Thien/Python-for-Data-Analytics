```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
#data cleanup
df['job_posted_date'] =pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:,'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
```

```python
# fill holes
median_salary_year = df['salary_year_avg'].median()
median_salary_hour = df['salary_hour_avg'].median()
```

```python
df_filled = df

df['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled.loc[:10,'salary_year_avg':'salary_hour_avg']
```

```python
# remove duplicates
df_unique = df_filled

df_unique = df_unique.drop_duplicates()

# shows how many duplicates were removed

print('Original df: ',len(df_filled))
print('Drop duplicates df: ',len(df_unique))
print('Row dropped: ',len(df_filled)-len(df_unique))
```

```python
# remove duplicates jobs from dif companies
df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])

# shows how many duplicates were removed
print('Original df: ',len(df_filled))
print('Drop duplicates df: ',len(df_unique))
print('Row dropped: ',len(df_filled)-len(df_unique))
```

```python
# Data Management
df.sample() #return random index
df.copy()#create a copy
```

```python
#Pivot table

df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc=('median'))
```

```python
top_countries = df['job_country'].value_counts().head(6).index

df_job_countries_salary = df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc=('median'))

df_job_countries_salary =df_job_countries_salary.loc[top_countries]

job_title =['Business Analyst','Cloud Engineer','Data Analyst']

df_job_countries_salary = df_job_countries_salary[job_title]

df_job_countries_salary.plot(kind='bar')

df_job_countries_salary
```
